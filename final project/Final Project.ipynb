```python
#import our data process library (note: you may have to install this)
import pandas as pd 

#lets use this to upload a sample dataset and show the starrt of the dataset
# Note that you need to download the dataset and make sure it's in the same 
# directory as your notebook
#this CSV was included in the final project upload
data= pd.read_csv("school-shootings-data.csv", encoding = 'unicode_escape')
data.head()
```

```python
data1 = data[['year','killed','injured','casualties','shooting_type']]

data2 = data[['year', 'city','state','killed','injured','casualties', 'school_type']]
```

```python
#Now let's visualize the data 
import altair as alt




# Let's implement filtering using dynamic queries. 
dropdown = alt.binding_select (options=["casualties", "killed", "injured"], name="Select a type of casualty:")

# Create a new selection that uses my dynamic query widget
selection = alt.selection(type="single", fields=['column'], bind=dropdown, init={'column':'casualties'})

# Let's specify our chart
alt.Chart(data).transform_fold(
    ["casualties", "killed", "injured"],
    as_=['column', 'People Impacted']
).transform_filter(
    selection
).mark_circle().encode(
    x=alt.X("year", bin = False, scale=alt.Scale(domain=[1999,2022])),
    y = "People Impacted:Q",
    color="shooting_type",
   # size="value:Q",
    tooltip=["shooting_type","casualties"],
).add_selection(selection)



#TASK 1
#SELECT A TYPE OF CASUALTY TO EXPLORE THE VISUALIZATION 
```

```python
selection = alt.selection(type="multi", fields=["state"])

base =  alt.Chart(data).properties(width=500, height=250)

overview = alt.Chart(data).mark_bar().encode(
    y = "casualties",
    x = "state",
    color=alt.condition(selection, alt.value("darkred"), alt.value("lightgrey"))
).add_selection(selection).properties(height=250, width=500)

detail = hist = base.mark_circle().encode(
    y = "casualties", 
    x = "city",
    color = alt.Color('school_type', scale=alt.Scale(scheme='category20'))
).transform_filter(selection).properties(height=250, width=250)

overview | detail


#TASK 2
#SELECT A STATE TO EXPLORE THE VISUALIZATION
```
